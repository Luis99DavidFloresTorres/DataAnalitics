Realizamos la función tomando en cuenta la variable Cilindros y con sus métodos respectivos realizar la fórmula, retornando valores tipo String es decir cualitativas con una lógica implementada.

```python
def variablesCualitativas(variable):
  alto = data['Cilindros'].mean()+(1*data['Cilindros'].std())
  bajo = data['Cilindros'].mean()-(1*data['Cilindros'].std())
  if variable > alto:
    return "alto"
  if variable < alto and variable > bajo:
    return "medio"
  if variable < bajo:
    return "bajo"
```

Aplicamos dicha función en la variable elegida

```python
data['Cilindros'].apply(variablesCualitativas)
```

Realizamos las graficas segun las variables cualitativas excepto en la curva(histplot). La variable ejercicio1 es la columna de paises codificada a números.

```python
def dibujarMapas():
  sns.barplot(x = 'Millas_Por_Galon',y='Pais_Origen',data = data)
  plt.show()
  sns.boxplot(x = 'Millas_Por_Galon',y='Pais_Origen',data = data)
  plt.show()
  sns.histplot(data=data, x="Millas_Por_Galon", kde=True)
  plt.show()
  sns.heatmap(data.corr()[['ejercicio1']].sort_values(by='ejercicio1', ascending=False), vmin=-1, vmax=1, annot=True, cmap='BrBG')
  plt.show()
```
